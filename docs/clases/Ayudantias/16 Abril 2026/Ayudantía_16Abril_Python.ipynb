### Modelos Parametricos

```python
import pandas as pd

# Dataset viene de: https://www.kaggle.com/datasets/souradippal/simple-linear-regression-hours-vs-marks-data

df = pd.read_csv("./students.csv")
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Hours_Studied'],df['Marks'], color = 'blue', marker = 'o')
plt.title("Datos")
plt.xlabel("Horas estudiadas")
plt.ylabel("calificación")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# modelo

model = LinearRegression()
model.fit(df[['Hours_Studied']],df['Marks'])
```

```python
import numpy as np

X_continuo = np.linspace(0,12,100).reshape(-1,1)

y_pred = model.predict(X_continuo)
```

```python
plt.plot(X_continuo, y_pred, color = "red")
plt.scatter(df['Hours_Studied'],df['Marks'], color = 'blue', marker = 'o')
plt.title("Modelo de regresió lineal")
plt.xlabel("Horas estudiadas")
plt.ylabel("calificación")
plt.show()
```

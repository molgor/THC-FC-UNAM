### Introducción a la Simulación

El **Método de Montecarlo** es una técnica que utiliza el **azar** para resolver problemas numéricos. Consiste en realizar miles de experimentos aleatorios (como lanzar dardos a una pared) para estimar un resultado final, como un área o una constante matemática.


#### Ejercicio 1: Estimación de $\pi$ (El dardo y el círculo)

Imagina que lanzas dardos aleatoriamente a un tablero cuadrado de lado $2$. Dentro del cuadrado hay un círculo de radio $1$. Si lanzas suficientes dardos, la proporción de aciertos nos permite calcular el valor de $\pi$ mediante la siguiente relación:

$$
\pi \approx 4 \times \frac{\text{Dardos dentro del círculo}}{\text{Total de dardos lanzados}}
$$

```python
import random
import matplotlib.pyplot as plt

class dardo:
    def __init__(self,x,y):
        self.x = x
        self.y = y
        self.adentro = (self.x**2 + self.y**2) <= 1

class persona:
    def __init__(self, nombre):
        self.nombre = nombre
        self.lanzamientos = []

    def lanzar_dardos(self, cantidad):
        for i in range(cantidad):
            x_azar = random.uniform(-1,1)
            y_azar = random.uniform(-1,1)
            nuevo_dardo = dardo(x_azar,y_azar)
            self.lanzamientos.append(nuevo_dardo)

    def calcular(self):
        if not self.lanzamientos: return 0
        aciertos = sum(1 for d in self.lanzamientos if d.adentro)
        pi = 4 * (aciertos/len(self.lanzamientos))
        return pi

    def graficar_resultados(self):
        x_in = [d.x for d in self.lanzamientos if d.adentro]
        y_in = [d.y for d in self.lanzamientos if d.adentro]
        x_out = [d.x for d in self.lanzamientos if not d.adentro]
        y_out = [d.y for d in self.lanzamientos if not d.adentro]

        plt.figure(figsize=(7,7))
        plt.scatter(x_in,y_in, color='blue', s=1, label = 'adentro')
        plt.scatter(x_out,y_out,color = 'red', s=1,label = 'afuera')
        circ = plt.Circle((0,0),1,color = 'black', fill=False, linewidth = 1)
        plt.gca().add_artist(circ)
        plt.title(f"{self.nombre} - Pi aproximado: {self.calcular(): .4f}")
        plt.axis('equal')
        plt.show()

estudiante = persona("Israel")
estudiante.lanzar_dardos(10000000)
estudiante.graficar_resultados()
```

De la misma manera, podemos usar "dardos" para calcular integrales (áreas bajo una curva). Si lanzamos puntos aleatorios en un cuadrado de $1 \times 1$, la cantidad de puntos que caen debajo de la función nos dará el área de la figura:

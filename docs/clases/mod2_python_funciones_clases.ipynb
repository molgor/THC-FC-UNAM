# Funciones y Clases en Python
## Estructura simple de una función

```python
def f():
    """
    Esta es una función abstracta

    Recibe: Nada
    Regresa: Saludo (String)
    """
    saludo = "Soy la función f"
    return saludo
    
```

```python
saludo = f()
```

```python
saludo
```

## Ejemplo de función numérica

```python
def f(x):
    """
    Esta es una función numérica abstracta

    Recibe: x : float / intero
    Regresa: función evaluada 
    """
    a = 0.4
    b = 1.0
    y = a*x**2 + b 
    return y
    
```

## Función con argumentos opcionales

```python
def f(x,a=0.5,b=1.0,c=-3.0):
    """
    Esta es una función que implementa una parábola

    Recibe: x : float / intero
            a : (opcional, default: 0.5)
            b : (opcional, default: 1.0)
            c : (opcional, default: -3)
    Regresa: función evaluada 
    """
    y = a*x**2 + c*x + b 
    return y
    
```

```python
f(2.3,a=1.0,b=2.0)
```

```python
import numpy as np
```

```python
%time xs = np.linspace(-3,3,10000)
```

## Dos formas para evaluar la función

1. Ciclo For normal


```python
%%time
ys = []
for x in xs:
    y = f(x)
    ys.append(y)
```

## Otra forma, a través de "List Comprehension" (compresión de listas)

```python
%time ys = [f(w) for w in xs]
```

## Otra forma, a través de funciones anónimas (lambda) y mapeos

```python
ys = list(map(f,xs))
```

```python
c = 1.0
nf = lambda x: x**2 + c
```

```python
%time ys = map(lambda x: x**2 +c, xs)
```

```python
%time ys = list(ys)
```

## Graficación
Vamos a usar `matplotlib`

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(xs,ys)
plt.title('Función parábola: a = 0.5,b=1.0') 
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
#plt.savefig('prueba_parabola.png')
```

## Múltiples gráficas

```python
import matplotlib as mpl
```

```python
# múltiples valores de parámetros
nuevo_c = 1.0
n_as = np.linspace(-5,5,10)
n_bs = np.linspace(-10, 10, 10)
xs = np.linspace(-30,30,10000)

cmap = mpl.colormaps['plasma']
colors = [cmap(i) for i in np.linspace(0, 1, 100)]

i = 0
for n_a in n_as:
    for n_b in n_bs:
        nys = [f(x,a=n_a,b=n_b,c=nuevo_c) for x in xs]
        plt.scatter(xs,nys,color=colors[i])
        i += 1 # Contador sobre los valores de color.
        
plt.title(f'Múltiples parábolas') 
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```
